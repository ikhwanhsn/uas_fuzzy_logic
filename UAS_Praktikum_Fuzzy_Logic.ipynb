```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
# Input
suhu = ctrl.Antecedent(np.arange(35, 41, 0.1), 'suhu')
trombosit = ctrl.Antecedent(np.arange(50, 250, 1), 'trombosit')
leukosit = ctrl.Antecedent(np.arange(2, 10, 0.1), 'leukosit')

# Output
risiko = ctrl.Consequent(np.arange(0, 101, 1), 'risiko')
```

```python
suhu['normal'] = fuzz.trimf(suhu.universe, [35, 36.5, 37.5])
suhu['demam'] = fuzz.trimf(suhu.universe, [37, 38, 39])
suhu['tinggi'] = fuzz.trimf(suhu.universe, [38.5, 39.5, 40])

trombosit['rendah'] = fuzz.trimf(trombosit.universe, [50, 100, 150])
trombosit['normal'] = fuzz.trimf(trombosit.universe, [120, 180, 250])

leukosit['rendah'] = fuzz.trimf(leukosit.universe, [2, 3.5, 4.5])
leukosit['normal'] = fuzz.trimf(leukosit.universe, [4, 6, 9])

risiko['rendah'] = fuzz.trimf(risiko.universe, [0, 25, 50])
risiko['sedang'] = fuzz.trimf(risiko.universe, [40, 60, 80])
risiko['tinggi'] = fuzz.trimf(risiko.universe, [70, 85, 100])
```

```python
rule1 = ctrl.Rule(suhu['tinggi'] & trombosit['rendah'] & leukosit['rendah'], risiko['tinggi'])
rule2 = ctrl.Rule(suhu['demam'] & trombosit['rendah'], risiko['tinggi'])
rule3 = ctrl.Rule(suhu['demam'] & trombosit['normal'], risiko['sedang'])
rule4 = ctrl.Rule(suhu['normal'] & trombosit['normal'] & leukosit['normal'], risiko['rendah'])
rule5 = ctrl.Rule(suhu['tinggi'] & leukosit['normal'], risiko['sedang'])
```

```python
risk_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
risk_sim = ctrl.ControlSystemSimulation(risk_ctrl)

risk_sim.input['suhu'] = 38.5
risk_sim.input['trombosit'] = 120
risk_sim.input['leukosit'] = 3.8

risk_sim.compute()
print("Nilai Risiko DBD:", risk_sim.output['risiko'])
```

CODING ANFIS – Prediksi Produksi Jagung

```python
import numpy as np

x1 = 60   # soil moisture
x2 = 26   # avg temperature

params = [
    (0.0692, 0.0461, 0.00231),
    (0.0698, 0.0390, 0.00139),
    (0.0585, 0.0183, 0.00084),
    (0.0505, 0.0318, 0.00115)
]
```

```python
outputs = []
for p, q, r in params:
    y = p * x1 + q * x2 + r
    outputs.append(y)

print("Output tiap rule:", outputs)
```

```python
final_output = np.mean(outputs)
print("Prediksi Produksi Jagung (ton/ha):", round(final_output, 2))
```
